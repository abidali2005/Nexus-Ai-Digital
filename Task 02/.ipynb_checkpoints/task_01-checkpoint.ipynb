```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings

warnings.filterwarnings('ignore')
```

# Load Data

```python
df_train = pd.read_csv('OnlineRetail.csv' , encoding='latin1')
```

# Data Cleaning

```python
df_train.head()
```

# Handling Null Values

```python
df_train.isnull().sum()
```

```python
df_train = df_train.dropna(subset=['Description'])
```

# Handling Duplicates

```python
df_train.duplicated().sum()
```

```python
df_train = df_train.drop_duplicates()
```

```python
df_train.duplicated().sum()
```

```python
df_train['InvoiceDate'] = pd.to_datetime(df_train['InvoiceDate'])
```

```python
df_train['Year'] = df_train['InvoiceDate'].dt.year
df_train['Month'] = df_train['InvoiceDate'].dt.month
df_train['Day'] = df_train['InvoiceDate'].dt.day
df_train['Hour'] = df_train['InvoiceDate'].dt.hour
df_train['DayOfWeek'] = df_train['InvoiceDate'].dt.day_name()
df_train['Quarter'] = df_train['InvoiceDate'].dt.quarter
```

```python
df_train = df_train.drop(columns='InvoiceDate' , axis=1)
```

```python
df_train['Sales'] = df_train['Quantity'] * df_train['UnitPrice']
```

# Descriptive Statistics

```python
df_train.describe()
```

# Visualization

```python
monthly_sales = df_train.groupby(['Year','Month'])['Sales'].sum().reset_index()
monthly_sales['YearMonth'] = pd.to_datetime(monthly_sales[['Year','Month']].assign(DAY=1))

plt.figure(figsize=(8,5))
sns.lineplot(data=monthly_sales, x='YearMonth', y='Sales', marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_10_countries = df_train.groupby('Country')['Sales'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
plt.title("Avg Sales of Top 10 Countries")
sns.barplot(top_10_countries)
plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
plt.title("Yearly Sales Trend")
sns.barplot(x=df_train['Year'] , y=df_train['Sales'])
plt.show()
```

```python
top_10_products = df_train.groupby('Description')['Sales'].sum().nlargest(10).reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x=top_10_products['Sales'] , y=top_10_products['Description'] , palette='viridis')
plt.title("High Saling Top 10 Products")
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5))
sales_by_day = df_train.groupby('DayOfWeek')['Sales'].sum()
sns.barplot(sales_by_day , palette='magma')
plt.title("Weekly Sales Trend")
plt.tight_layout()
```

```python
hourly_sales = df_train.groupby('Hour')['Sales'].mean()
plt.figure(figsize=(8,5))
plt.title("Hourly Sales Trend")
sns.lineplot(hourly_sales , marker = 'o' , color = 'red')
```

```python
sales_in_quarter = df_train.groupby('Quarter')['Sales'].sum()
plt.figure(figsize=(8,5))
sns.barplot(sales_in_quarter , palette='magma')
plt.title("Quarter Wise Sales Trend")
plt.tight_layout()
```

```python
avg_quantity_in_country = df_train.groupby('Country')['Quantity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(avg_quantity_in_country , palette='viridis')
plt.title(" Top 10 Avg Quantity Sale in Country")
plt.tight_layout()
```

```python
top_stock = df_train.groupby('StockCode')['Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=top_stock, x='Sales', y='StockCode', palette="coolwarm")
plt.title("Top 10 Stocks")
plt.show()
```

```python
# Top StockCode with Description for clarity
top_stock_desc = (df_train.groupby(['StockCode','Description'])['Sales']
                  .sum()
                  .nlargest(10)
                  .reset_index())

plt.figure(figsize=(12,6))
sns.barplot(data=top_stock_desc, x='Sales', y='StockCode', hue='Description', dodge=False)
plt.title("Top 10 Stocks related to products")
plt.show()
```

```python

customer_sales = df_train.groupby('CustomerID')['Sales'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.histplot(customer_sales[customer_sales['Sales'] < 10000]['Sales'], 
             bins=20, kde=True, color="blue")
plt.title("Histogram of Customer Sales (< 10,000)")
plt.show()

```

```python
df_train.to_csv('cleaned.csv' , index=False)
```
